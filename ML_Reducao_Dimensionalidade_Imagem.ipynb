```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import mahotas
```

```python
img=cv2.imread('kindpng_850924.png')
img_res = cv2.resize(img, dsize=(490, 450), interpolation=cv2.INTER_LINEAR) 

cv2_imshow(img_res)

img_res=cv2.cvtColor(img_res,cv2.COLOR_BGR2GRAY)

cv2_imshow(img_res)
```

```python
blur = cv2.GaussianBlur(img_res, (7, 7), 0)
```

```python
#Binarização em pixels brancos e pretos
T = mahotas.thresholding.otsu(blur)
bin = blur.copy()
bin[bin > T] = 255
bin[bin < 255] = 0
bin = cv2.bitwise_not(bin)
cv2_imshow(bin)
```
